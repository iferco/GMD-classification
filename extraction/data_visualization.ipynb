```python
#Coordenadas
#"Lugo, A Fonsagrada"
def get_coordinates(place):
    geolocator = Nominatim(user_agent="galician_music")
    location = geolocator.geocode(place)
    if location is None:
        return 0, 0
    return location.latitude, location.longitude

print(get_coordinates("Laxe, A Coruña"))
```

```python

# add new column to dataframe
df['classification'] = "Not classified"

#Now function that receives coords and itereates through the dataframe, adding a new classification 
#column
def get_muiñeiras_in_area(coords,type):
    
    #add new column to dataframe
    for index, row in df.iterrows():
        #Get coordinates from dataframe
        lat = row['lat']
        lon = row['lon']
        #Check if coordinates are inside the area
        if coords[0][0] > lat > coords[2][0] and coords[0][1] < lon < coords[1][1]:
            #If they are, append the row to the new dataframe
            df.loc[index,'classification'] = type

            

get_muiñeiras_in_area(coords_costa_do_morte, "Costa do Morte")
get_muiñeiras_in_area(coords_ancares, "Ancares")
get_muiñeiras_in_area(coords_courel, "Courel")
get_muiñeiras_in_area(coords_rias_baixas, "Rias Baixas")
get_muiñeiras_in_area(coords_a_coruña, "A Coruña")
get_muiñeiras_in_area(coords_fonsagrada, "A Fonsagrada-Asturias")
get_muiñeiras_in_area(coord_portuguese_frontier, "Portuguese Border")


```
